```python
import cv2
import csv
```

# Get metadata of training videos

```python
index = 0
indexlimit = 104
filepath = str('C:\\Users\\Asus\\OneDrive\\Studium\\Semester 7\\Deep Learning\\Projekt Videos\\Task 1 Training\\Task1_Training_Videos')

# creating a new .csv file
with open('training_data_exploration.csv', 'w', newline='') as file:
        writer = csv.writer(file, delimiter=";")
        writer.writerow(["Videoname", "fps", "total_frames", "duration in seconds", "width", "height"])
        
print("Created .csv successfully")
        
while index <= indexlimit:
    
    # build the correct filename
    if index <10:
        filename = "\\_8Vy3dlHg2w_0000" +str(index) +".mp4"
    elif index >= 10 and index <100:
        filename = "\\_8Vy3dlHg2w_000" +str(index) +".mp4"
    elif index >=100:
        filename = "\\_8Vy3dlHg2w_00" +str(index) +".mp4"
        
    # build the correct path
    fullpath = filepath + filename
    #print(fullpath)
    
    # load the current video
    cap = cv2.VideoCapture(fullpath)
    
    # get the needed parameter
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_count/fps
    duration = str(duration)
    width  = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
    height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    
    # set the parameters of the current video in an array
    values = [filename, fps, frame_count, float(duration), width, height]
    
    # append the parameters to the csv file
    with open('training_data_exploration_WH.csv','a') as fd:
        writer = csv.writer(fd, delimiter=";")
        writer.writerow(values)
    
    print("processed video " +str(index) +"/" +str(indexlimit))
    
    index += 1

print("Successfully filled .csv file with metadata.")
```
